```python
import os
import glob
import numpy as np
import json
from func import *
import pandas as pd
```

```python
opt_folder_path = './data/cxr_opt_healthcheck_zfill3'
opt_user_name = ['u2_u11', 'u3_u12', 'u4_u13', 'u5_u14', 'u6_u15', 'u7_u16', 'u8_u17', 'u9_u18', 'u10_u19']
use_ais = ['without_AI', 'with_AI']
```

```python
mapping_df = pd.read_csv('./data/brmh_2_mapping_table_healthcheck.csv')
mapping_cases = mapping_df['case_no'].tolist()
```

```python
df = pd.DataFrame(columns = ['opt_user', 'case_name', 'seq','label_without', 'rate_without', 'label_with', 'rate_with'])

for i in range(len(opt_user_name)):
#     if i == 1:
#         break
#     i=5
    # without_AI
    print(opt_user_name[i], use_ais[0])
    opt_without_path = os.path.join(opt_folder_path,opt_user_name[i],use_ais[0])
    opt_without_json_li = glob.glob(opt_without_path+'/*.json')
    
    opt_with_path = os.path.join(opt_folder_path,opt_user_name[i],use_ais[1])
    opt_with_json_li = glob.glob(opt_with_path+'/*.json')

    for k in range(len(opt_without_json_li)):
        opt_without = opt_without_json_li[k]
        contours_opt_without, contourIds_without, rating_list_without, rate_without = json_opt_open(opt_without)
        name_without = opt_without[opt_without.rindex('\\')+1:-5].zfill(3)
        
        opt_with = opt_with_json_li[k]
        contours_opt_with, contourIds_with, rating_list_with, rate_with = json_opt_open(opt_with)
        name_with = opt_with[opt_with.rindex('\\')+1:-5].zfill(3)
        
        case_name = mapping_cases[int(name_without)-1]
        
        if name_with==name_without:
            if contours_opt_without:
                label_without = 1
            else:
                label_without = 0

            if contours_opt_with:
                label_with = 1
            else:
                label_with = 0
        else:
            print('no match')
        
#         print(opt_user_name[i], name_without, label_without, rate_without, label_with, rate_with)
        data = {'opt_user':opt_user_name[i], 'case_name':case_name, 'seq':name_without, 'label_without':label_without, 'rate_without':rate_without, 
                'label_with':label_with, 'rate_with':rate_with}
        
        df=df.append(data , ignore_index=True)
```
