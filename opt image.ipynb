```python
import os
import json
import numpy as np
import ast
import cv2
from func import *
```

```python
t1_path = './data/cxr_opt_respiratory'
t1_u_name = os.listdir(t1_path)
# print(t1_u_name)

gt_path = './data/review_result_opt/respiratory/'
AIs = ['without_AI','with_AI']

for AI in AIs:
    print(AI)
    for i in range(len(t1_u_name)):
#         if i==1:
#             break

        t1_json_path = os.path.join(t1_path, t1_u_name[i], AI)
        t1_json_file = os.listdir(t1_json_path)

        for k in range(len(t1_json_file)):
#             if k==14:
#                 break

            #########################
            #######  gt
            #########################
            with open(gt_path+'/'+t1_json_file[k], "r") as st_json:
                file_gs = json.load(st_json)
                keys_gs = list(file_gs.keys())
                name = t1_json_file[k][:-5]
                if 'abnormal_finding' in keys_gs:
        #                     print(len(file_gs['abnormal_finding']))
                    if file_gs['abnormal_finding'] == [] :
                        label = 'normal'
                        width_gs = file_gs['width']
                        height_gs = file_gs['height']

                    else:
                        for j in range(len(file_gs['abnormal_finding'])):
                            roi_gs = file_gs['abnormal_finding'][j]['contour_list']
                            keys = list(roi_gs.keys())
                            for m in range(len(keys)):
                                roi_gs = file_gs['abnormal_finding'][j]['contour_list']
                                keys = list(roi_gs.keys())
    #                             print(len(keys),m)
                                roi_gs = roi_gs[keys[m]]
                                label = file_gs['abnormal_finding'][j]['label_text']
                                width_gs = file_gs['width']
                                height_gs = file_gs['height']
                                xy = []
                                for roi in roi_gs:
                                    x = roi['x']+int(width_gs/2)
                                    y = roi['y']+int(height_gs/2)
                #                     x = abs(roi['x'])
                #                     y = abs(roi['y'])
                                    xy.append((x,y))
                                xy = np.array(xy)

                                img_gs_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'gt',t1_u_name[i])
                                CreateFolder(img_gs_path)

                                if j==0 and m==0:
                                    img_bg = np.zeros((height_gs,width_gs))
                                    image = cv2.fillPoly(img_bg, [xy], (255, 255, 255))
        #                                 img_gs_path = os.path.join('D:/lunit/data/img','result',ai_all_file[i],name+'dcm')
        #                                 CreateFolder(img_gs_path)
                                    cv2.imwrite(img_gs_path+'/'+name+'_gt.png', image)

                                    img_total_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'total',t1_u_name[i],name)
                                    CreateFolder(img_total_path)
                                    cv2.imwrite(img_total_path+'/'+name+'_gt.png', image)

                                else:
                                    print(name)
                                    img_bg = cv2.imread(img_gs_path+'/'+name+'_gt.png')
                                    image = cv2.fillPoly(img_bg, [xy], (255, 255, 255))
        #                                 img_gs_path = os.path.join('D:/lunit/data/img','result',ai_all_file[i],name+'dcm')
        #                                 CreateFolder(img_gs_path)
                                    cv2.imwrite(img_gs_path+'/'+name+'_gt.png', image)

                                    img_total_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'total',t1_u_name[i],name)
                                    CreateFolder(img_total_path)
                                    cv2.imwrite(img_total_path+'/'+name+'_gt.png', image)

                else:
                    label = 'normal'
                    width_gs = file_gs['width']
                    height_gs = file_gs['height']

            if label == 'normal':
                img = np.zeros((height_gs,width_gs))
                img_gs_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'gt',t1_u_name[i])
                CreateFolder(img_gs_path)
                cv2.imwrite(img_gs_path+'/'+name+'_gt.png', img)

                img_total_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'total',t1_u_name[i],name)
                CreateFolder(img_total_path)
                cv2.imwrite(img_total_path+'/'+name+'_gt.png',img)

            ##########################
            ########## opt(n)
            ##########################
            with open(t1_json_path+'/'+t1_json_file[k], "r") as st_json:
                file_opt = json.load(st_json)
                contour_dict=ast.literal_eval(file_opt['contour_list'])

                if contour_dict ==[]:
                    print('normal')
                    img = np.zeros((height_gs,width_gs))
                    img_gs_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'opt',t1_u_name[i])
                    CreateFolder(img_gs_path)
                    cv2.imwrite(img_gs_path+'/'+name+'_opt.png', img)

                    img_total_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'total',t1_u_name[i],name)
                    CreateFolder(img_total_path)
                    cv2.imwrite(img_total_path+'/'+name+'_opt.png',img)
                    
                else:
                    keys = list(contour_dict.keys())
                    print(keys)

                    for n in range(len(keys)):
                        contour = contour_dict[keys[n]]
                        arr = convert_dict_to_array(contour)
                        arr = arr*(width_gs,height_gs)
                        array = np.array(arr, dtype = 'int32')

                        img_gs_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'opt',t1_u_name[i])
                        CreateFolder(img_gs_path)

                        if n==0:
                            img_bg = np.zeros((height_gs,width_gs))
                            image = cv2.fillPoly(img_bg, [array], (255, 255, 255))
    #                                 img_gs_path = os.path.join('D:/lunit/data/img','result',ai_all_file[i],name+'dcm')
    #                                 CreateFolder(img_gs_path)
                            cv2.imwrite(img_gs_path+'/'+name+'_opt.png', image)

                            img_total_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'total',t1_u_name[i],name)
                            CreateFolder(img_total_path)
                            cv2.imwrite(img_total_path+'/'+name+'_opt.png', image)

                        else:
                            print(name)
                            img_bg = cv2.imread(img_gs_path+'/'+name+'_opt.png')
                            image = cv2.fillPoly(img_bg, [array], (255, 255, 255))
    #                                 img_gs_path = os.path.join('D:/lunit/data/img','result',ai_all_file[i],name+'dcm')
    #                                 CreateFolder(img_gs_path)
                            cv2.imwrite(img_gs_path+'/'+name+'_opt.png', image)

                            img_total_path = os.path.join('D:/lunit/img_opt_respiratory',AI,'total',t1_u_name[i],name)
                            CreateFolder(img_total_path)
                            cv2.imwrite(img_total_path+'/'+name+'_opt.png', image)
```
